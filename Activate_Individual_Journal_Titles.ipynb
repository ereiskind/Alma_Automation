# Activate Individual Journal Titles

## 1. Activate Title in IZ
### Method 1: Place Order Through Alma
### Method 2: Locate in CZ
1. Search electronic collections in CZ
2. Find the collection for the portfolio that should be activated
3. Click "Portfolio List" in the above collection's entry
4. Locate portfolio for the given title
5. Click "Activate" in the above portfolio's entry
6. Click "Activate" in the pop-up that appears without choosing a library

## 2. Activate Proxy for Electronic Collection
1. Search for electronic collection in IZ
2. Click into collection, then "Additional" tab
3. Go to Services and select "Edit" in menu
4. Go to "Linking" tab
5. Select "Yes" for "Proxy enabled"
6. Select "Defaut" for "Proxy selected"
7. Click "Save"
