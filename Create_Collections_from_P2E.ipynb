# Establish Collections Linked to CZ from P2E Portfolios

## 1. Create set containing the collection's portfolios
- Use advanced search with "Is standalone equals Yes" and filters to isolate the specific portfolios for the collection
  - Useful filtering criteria to use include the URL and interface (856$3) fields
- If multiple searches are needed to get all the portfolios, save each search as a set then run a combine sets job with the Boolean OR to create the final set
  - Alternatively, the "Add from Set" job can be triggered multiple times

## 2. Create the local electronic collection
1. In the menu, Resources > Add electronic local collection
2. Add a collection name to the "Public name" and "Public name (override)" fields
   - When the collection is linked to a CZ collection, the CZ collection name will override the value added to the "Public name" field
   - The collection name in "Public name (override)" will appear as the first line in the "full text availability" link box in Primo
3. "Description" field--don't see where its value appears in Primo; should it be used?
4. Add description *starting with "Perpetual access titles"* to the "Internal description" field
   - The fixed starting text can be used in an "Internal Description (Electronic Collection) | Starts With | Perpetual access titles" advanced search to find both collections created via this procedure and the portfolios contained within those collections
5. Set the "Library" field to "FSU Electronic Resource"
6. Click "Save and continue"
7. Switch "Service activation status" to "Available" **Might need to click "Save" at this point--did this at initial activation, but didn't work**
8. On the Portfolios tab
   1. Add > Add from Set
   2. Select the set created in step one
   3. A job will start for adding all the portfolios in the set to the collection
9. Click "Save"
10. Select the interface in the "Interface name" field
11. If all the portfolios were purchased together, select that PO line in the "PO Line" field
12. If all the portfolios are covered by the same license, select that license in the "License" field
13. On the Additional tab
    1. Add your name to the "Creator" field
    2. Make the appropriate selection for the "Is free?" field
14. Click "Save"
