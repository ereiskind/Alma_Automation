#### The cell.cross Template Document
* A .txt file
* What comes out of OpenRefine as an escaped double quote (`\"`) needs to have both the escape character and the double quote escaped, making the final metacharacter appear as `\\\"`
* Curly braces need to be escaped through duplication--where a standard JSON uses a curly brace, two facing in the given direction are needed
* The name of the project must be replaced with `{other_project}`
