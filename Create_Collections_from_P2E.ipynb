# Establish Collections Linked to CZ from P2E Portfolios

1.	Search electronic portfolios to get items for collection

    a.	Is standalone=yes

    b.	Other filtering criteria for the collection (interface, URL)
    
2.	Save search result as set

3.	Resource > Add electronic local collection

    a.	In creation wizard, portfolios tab > add titles from set

        i.	select set created above

        ii.	This addition is done as a job

    b.	In the “Public Name (Override)” field, add collection name suitable for use as the first line in the “full text availability” link box

    c.	Add further info to the “Description (Override)” field—**check if this is visible to public**

        i.	When the collection is linked to a CZ collection, the “Public name” and “Description” fields are inherited from the linked collection

    d.	**Add some sort of marker for perpetual access collections**

4.	Search for collections

    a.	CZ collection with holdings matching local collection

    b.	newly created collection

5.	In new collection search result option, in the extra options, select “Connect to Community”

    a.	Find matching CZ collection selected above

    b.	Check no on “Use the community bibliographic records”

    c.	**CLARIFY: CZ collection will need to be activated (part of full text service)**
    
    d.	Linking collections run as job
